```
import matplotlib.pyplot as plt
```

```
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.plot(x,y)
```

```
fig1 = plt.figure()

ax1 = fig1.add_axes([0,0,1,1])
ax2 = fig1.add_axes([0.2,0.5,.2,.2])


ax1 = fig1.add_axes([0,0,1,1])
ax2 = fig1.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y,'r')
ax2.plot(x,y,'r')
```

```
fig1 = plt.figure()
ax0 = fig1.add_axes([0,0,1,1])
ax1 = fig1.add_axes([0.2,0.5,.4,.4])

ax0.set_xlabel('X')
ax0.set_ylabel('Z')

ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_ylim(30,50)
ax1.set_xlim(20,22)

ax0.plot(x,z)
ax1.plot(x,y)
fig1 = plt.figure()
ax1.set_title('zoom')
ax3 = fig1.add_axes([0,0,1,1])
ax4 = fig1.add_axes([0.2,0.5,.4,.4])
```

```
fig,axes = plt.subplots(nrows=1,ncols=2)
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,color='red',lw="5",ls="--")
axes[1].plot(x,z,color='blue',lw="3")
fig,axes = plt.subplots(nrows=1,ncols=2, figsize=(8,2))

axes[0].plot(x,y,color='red',lw="5",ls="--")
axes[1].plot(x,z,color='blue',lw="3")
```
